```python
dbutils.widgets.dropdown("env", "dev", ["dev", "qa", "prod"])
env = dbutils.widgets.get("env")

config = get_env_config(env)

catalog = config["catalog"]
schema = config["schema"]
```

```python
raw_path_orders = f"/Volumes/{catalog}/{schema}/raw_volume/orders"
raw_path_customers = f"/Volumes/{catalog}/{schema}/raw_volume/customers"
bronze_path = f"/Volumes/{catalog}/{schema}/bronze_volume/"
```

```python
customers_df = spark.read.option("header", True).option("inferSchema", "true").csv(raw_path_customers)
orders_df = spark.read.option("header", True).option("inferSchema", "true").csv(raw_path_orders)
```

```python
table_name = f"{catalog}.{schema}.bronze_customers"

customers_df.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable(table_name)
```
